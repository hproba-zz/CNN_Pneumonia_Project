```python
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
# Initialising the CNN
cnn = Sequential()
```

```python
# Step 1 - Convolution
cnn.add(Convolution2D(32, 3, 3, input_shape = (64, 64, 3), activation='relu'))
```

```python
# Step 2 - Pooling
cnn.add(MaxPooling2D(pool_size=(2,2)))
```

```python
# Adding a second convolutional layer
cnn.add(Convolution2D(32, 3, 3, activation='relu'))
cnn.add(MaxPooling2D(pool_size=(2,2)))
```

```python
# Step 3 - Flattening
cnn.add(Flatten())
```

```python
# Step 4 - Full connection
cnn.add(Dense(output_dim=128, activation='relu'))
cnn.add(Dense(output_dim = 1, activation='sigmoid'))
```

```python
# Compiling the CNN
cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Part 2 - Fitting the CNN to the images
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                  shear_range=0.2,
                                  zoom_range=0.2,
                                  horizontal_flip=True)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
training_set = train_datagen.flow_from_directory('../input/chest_xray/chest_xray/train',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')

test_set = test_datagen.flow_from_directory('../input/chest_xray/chest_xray/test',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')

cnn.fit_generator(training_set,
                         samples_per_epoch = 8000,
                         nb_epoch = 10,
                         validation_data = test_set,
                         nb_val_samples = 2000)
```
